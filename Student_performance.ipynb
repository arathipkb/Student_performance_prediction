```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df.info()
df.describe()
```

```python
print("Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
df["Hours_Studied"].fillna(df["Hours_Studied"].mean())
df["Attendance"] = df["Attendance"].fillna(df["Attendance"].mean())
df["Exam_Score"] = df["Exam_Score"].fillna(df["Exam_Score"].mean())
```

```python
df = df.drop_duplicates()
```

```python
print("After Removing Duplicates:")
df.head()
```

```python
df["Parental_Education_Level"] = df["Parental_Education_Level"].str.upper()
df["Access_to_Resources"] = df["Access_to_Resources"].str.upper()
df["Extracurricular_Activities"] = df["Extracurricular_Activities"].str.upper()
```

```python
print("After Text Cleaning:")
df.head()
```

```python
df.loc[df["Hours_Studied"] > 24, "Hours_Studied"] = df["Hours_Studied"].mean()
df.loc[df["Attendance"] > 100, "Attendance"] = df["Attendance"].mean()
```

```python
print("After Outlier Removal:")
df.head()
```

```python
df["Pass"] = df["Exam_Score"].apply(lambda x: 1 if x >= 40 else 0)
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```
